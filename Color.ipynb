```python
#Morphological operators
img = cv2.imread("Morphology_1_Tutorial_Theory_Dilatation_2.jpg",0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
kernel = np.ones((5,5))
```

```python
#Erosion
erode = cv2.erode(img,kernel,iterations=1)
```

```python
plt.imshow(erode,cmap='gray')
```

```python
#Dilation
dilate = cv2.dilate( img,kernel,iterations=1)
```

```python
plt.imshow(dilate,cmap='gray')
```

```python
#Opening
op = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(op,cmap='gray')
```

```python
#Closing
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(closing,cmap='gray')
```

```python
#Gradient
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient,cmap='gray')
```

```python
cross =cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))

temp = cv2.morphologyEx(img,cv2.MORPH_CLOSE,cross)
plt.imshow(temp,cmap='gray')
```
